```python
import requests
```

```python
url_github = 'https://api.github.com/search/repositories?q=language:python&sort=stars'
headers_github = {"Accept": 'application/vnd.github.v3+json'}

r_github = requests.get(url_github, headers = headers_github)

print(f"Status Code:{r_github.status_code}")
```

```python
response_dict = r_github.json()
response_dict
```

```python
response_dict.keys()
```

```python
print(f"Total Repositories:{response_dict['total_count']}")
```

```python
repo_dicts = response_dict['items']
print(f"Repositories returned: {len(repo_dicts)}")
```

```python
repo_dict = repo_dicts[0]
print(f"\nKeys:{len(repo_dict)}\n")

for key in sorted(repo_dict.keys()):
    print(key)
```

```python
print("Selected Information about first repository:")
print(f"Name:{repo_dict['name']}")
print(f"Owner:{repo_dict['owner']['login']}")
print(f"Description:{repo_dict['description']}")
# stars, repository, created, updated
```

```python
print("Selected Information about each repository:")


for repo in repo_dicts:
    
    print(f"Name:{repo['name']}")
    print(f"Owner:{repo['owner']['login']}")
    print(f"Description:{repo['description']}")
    print()
```

### Visualizing Repositories using Plotly

```python
from plotly.graph_objs import Bar
from plotly import offline

repo_names, stars = [], []
labels = []
repo_links = []

for repo in repo_dicts:
    repo_name = repo['name']
    repo_names.append(repo['name'])
    stars.append(repo['stargazers_count'])
    
    owner = repo['owner']['login']
    description = repo['description']
    label = f"{owner} <br />{description}"
    labels.append(label)
    
    repo_url = repo['html_url']
    repo_link = f"<a href='{repo_url}'>{repo_name}</a>"
    repo_links.append(repo_link)
print(stars)
print(repo_names)
print(labels)
print(repo_links)
```

```python
data = [{'type':'bar',
         'x':repo_links, 
         'y':stars,
         'hovertext':labels,
         'marker':{
             'color':'rgb(60,100,150)',
             'line':{'width':1.5, 'color':'rgb(25,25,25)'}
         },
         'opacity':0.6
        
        }]
my_layout = {'title':"Most-Starred Python Projects on Github",
             'xaxis':{'title':'Repository'},
             'yaxis':{'title':'Stars'}}
fig = {'data':data, 'layout':my_layout}
offline.plot(fig, filename='python_repos.html')
```

# The Hacker News API

```python
import requests
import json

url_hacker = 'https://hacker-news.firebaseio.com/v0/item/19155826.json'

r_hacker = requests.get(url_hacker)

response_hacker = r_hacker.json()

readable_file = "readable_hacker_news.json"
with open(readable_file, 'w') as f:
    json.dump(response_hacker, f, indent=4)
```

```python
from operator import itemgetter
url_topstories = "https://hacker-news.firebaseio.com/v0/topstories.json"
r_topstories = requests.get(url_topstories)

submission_ids = r_topstories.json()
submission_dicts = []

for submission_id in submission_ids[:30]:
    url = f"https://hacker-news.firebaseio.com/v0/item/{submission_id}.json"
    r = requests.get(url)
    response_hacker_dict = r.json()
    
    try:
        submission_dict = {
        'title':response_hacker_dict['title'],
        'hn_link':f"http://news.ycombinator.com/item?id={submission_id}",
        'comments':response_hacker_dict['descendants']
    }
    except KeyError:
        continue
    
    submission_dicts.append(submission_dict)

submission_dicts = sorted(submission_dicts, key=itemgetter("comments"), reverse = True)
```

```python
for submissions_dict in submission_dicts:
    print(f"\nTitle: {submissions_dict['title']}")
    print(f"Discussion Link:{submissions_dict['hn_link']}")
    print(f"Comments:{submissions_dict['comments']}")
```
